```python
import lightgbm as lgb
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from randomizer import ct

data = load_iris(as_frame=True)

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2)

# Create and train LightGBM model
model = make_pipeline(ct, lgb.LGBMClassifier())
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test[:1], pred_leaf=True)
prediction
```

```python
import mlflow
import mlflow.sklearn

# Save the model
with mlflow.start_run():
    model = mlflow.sklearn.log_model(model, "model")
```

```python
# Load the model
loaded_model = mlflow.sklearn.load_model(f"runs:/{model.run_id}/model")
```

```python
loaded_model.predict(X_test, pred_leaf=True)
```
